```python
def display_all(df):
    with pd.option_context("display.max_rows", 1000, "display.max_columns", 1000): 
        display(df)
```

```python
def proximityMatrix(model, X, normalize=True):      

    terminals = model.apply(X)
    nTrees = terminals.shape[1]

    a = terminals[:,0]
    proxMat = 1*np.equal.outer(a, a)

    for i in range(1, nTrees):
        a = terminals[:,i]
        proxMat += 1*np.equal.outer(a, a)

    if normalize:
        proxMat = proxMat / nTrees

    return proxMat   
```

```python
def readAndSavefile(PATH,filename,ram=0):
    df_raw = pd.read_csv(f'{PATH}{filename}', low_memory=False,sep=";")
    if (ram == 1):
        os.makedirs('tmp', exist_ok=True)
        df_raw.to_feather(f'tmp/{filename}-raw')
    return df_raw
```

```python
def statis(df_raw):    
    s1 = display_all(df_raw.tail().T)
    #statistics
    s2 = display_all(df_raw.describe(include='all').T)
    #missing calculator
    s3 = display_all(df_raw.isnull().sum().sort_index()/len(df_raw))
    
    return s1,s2,s3
```

```python
def forestCaller(df_raw,label):
    m = RandomForestClassifier(n_jobs=-1) 
    m.fit(df_raw,label)
    return m,m.score(df_raw,label)
```

```python
def format_dataset(df_raw,labelname):
    df = df_raw.drop(labelname, axis=1)
    y = df_raw.labelname
    return df,y
```

```python
class Disti:
    def __init__(self, values, counting):
        self.values = values
        self.counting = counting
```

```python
def synthetic_MarginalData(df_raw):
    columnsN=df_raw.columns
    list = []
    for colind in range(len(columnsN)):
        a = df_raw[columnsN[colind]].value_counts().axes
        b = df_raw[columnsN[colind]].value_counts().array
        list.append(Disti(a,b))
    arrRan=np.zeros(len(columnsN)-1)

    
    columns = columnsN
    df = pd.DataFrame(columns=columns)

    for i in range(len(df_raw)):
        for colind in range(len(columnsN)-1):
            arrRan[colind]=random.choices(np.reshape(np.array(list[colind].values),(np.shape(list[colind].values)[1],)),weights=np.array(list[colind].counting))[0]
        arr = np.append(arrRan,0)
        df.loc[i]= (arr)
    return df
    
```

```python
def makeDataset_UnSupervRF(df_raw,labelname):
    df_raw[[labelname]]=1
    new_df=pd.concat([df_raw , synthetic_MarginalData(df_raw)])
    return new_df
```

# Semi Supervision
1. Generate Sample Dataset
2. Assigned removed label Nan value
3. Generate Classs Each Tree with leafs


```python
class Nodes:
    def __init__(self,leafs,indleaf):
        self.leafs = leafs
        self.indleaf =indleaf
```

```python
class SemiDf:
    def __init__(self,data,label):
        self.data = data
        self.label = label
```

```python

class Treey:
    def __init__(self,index,nodes,weights):
        self.index = index
        self.nodes = nodes
        self.weights=weights
```

```python
def semiData_Gen(filepath,filename,labelname,dataRatio=0.7):
    semi_df=readAndSavefile(filepath, filename)
    num = np.random.choice(range(0, len(semi_df)),int(len(semi_df)*dataRatio),replace=False)
    if(type(semi_df[labelname][0]) is str):
        train_cats(semi_df)
        semi_df[labelname] = semi_df[labelname].cat.codes
    semi_df.at[num, labelname] = float("NaN")
    lbl = semi_df[labelname]
    return semi_df.drop(labelname,axis=1),lbl
```

```python
def leaf_Weight_helper(RF,SmDt): #random forest and semisupervised data
    terminals=RF.apply(SmDt.data)
    nTrees = terminals.shape[1]
    leafs = []
    storeNodes= []
    for i in range(nTrees):
        uniqleaf = set(terminals[:,i])
        leafs.append(uniqleaf)
        mylist = []
        for j in uniqleaf:
            l=[index for index, value in enumerate(terminals[:,i]) if value == j]
            mylist.append(l)
        n = Nodes(uniqleaf,mylist)
        storeNodes.append(n)
        mylist=[]
    return storeNodes
    
```

```python
def Tree_leaf_weight(y,SmDt):  
    mylist3 =[]
    len(y)
    weights= []
    listTree = []
    for j in range(len(y)):
        for i in y[j].indleaf:
            mylist3.append(SmDt.label[i])
        for i in mylist3:
            t=np.var(i)
            if(math.isnan(t)):
                t = 0.5
            weights.append(1-t)
        t=Treey(j,y[j],weights)
        listTree.append(t)
        weights=[]
        mylist3=[]
    return listTree
```

```python
def get_weighted_prm(tl,pmt):  
    for ti in range(len(tl)):
        for i in range(len(tl[ti].nodes.indleaf)):
            if(len(tl[ti].nodes.indleaf[i])>1):
                for j in range(len(tl[ti].nodes.indleaf[i])):
                    for k in range(j+1,len(tl[ti].nodes.indleaf[i])):
                        inj= tl[ti].nodes.indleaf[i][j]
                        ink= tl[ti].nodes.indleaf[i][k]
                        pmt[inj][ink]=pmt[inj][ink]+ tl[ti].weights[i]
                        pmt[ink][inj]=pmt[ink][inj]+ tl[ti].weights[i]
            else:
                pmt[tl[ti].nodes.indleaf[i],tl[ti].nodes.indleaf[i]]= 0
    return pmt
```

```python
def dataPrep(dr, arr):    
    semi_df = []
    lbl = []
    SmDt = []

    for val in arr:
        semi_d,lb = semiData_Gen("../../../datasets/",val,"y",dataRatio=dr)
        SmD=SemiDf(semi_d,lb)
        semi_df.append(semi_d)
        lbl.append(lb)
        SmDt.append(SmD)
    return SmDt,lbl
```

```python
def LeafWeightage(RF,SmDt,len_of_c_df):    
    y = []
    tl = []
    for i in range(len_of_c_df):
        y.append(leaf_Weight_helper(RF[i],SmDt[i])) #y[0].leafs #y[1].indleaf
        tl.append(Tree_leaf_weight(y[i],SmDt[i]))
    return tl
```

```python
def getProx(lbl,tl,len_ofc_df): 
    pmt = []
    pmt2 = []
    for i in range(len_ofc_df):
        pmt.append(np.zeros((len(lbl[i]), len(lbl[i]))))
        pmt2.append(get_weighted_prm(tl[i],pmt[i]))
        pmt[i] = (1 - pmt2[i])
    return pmt
```

```python
def get_clus(pmt,len_of_cdf):
    unsupK = []
    unsupC = []
    unsupS = []
    unsupSpec = []
    #unsupAp = []
    unsupH = []




    cluster_p = []
    cluster_c = []
    cluster_s = []
    cluster_spec = []
    #cluster_ap = []
    cluster_h = []


    
    for i in range(len_of_cdf):
        #print("dataset : " ,i )
        cluster_p.append(KMedoids(n_clusters=clus[i], random_state=0))
        cluster_p[i].fit_predict(pmt[i])
        cluster_p[i].labels_=cluster_p[i].labels_+1
        unsupK.append(sklearn.metrics.adjusted_rand_score(pred_actual[i], cluster_p[i].labels_))
        
        cluster_c.append(AgglomerativeClustering(n_clusters=clus[i],
                                  affinity='precomputed',
                                  linkage='complete'))
        cluster_c[i].fit_predict(pmt[i])
        cluster_c[i].labels_=cluster_c[i].labels_+1
        unsupC.append(sklearn.metrics.adjusted_rand_score(pred_actual[i], cluster_c[i].labels_))
    
        cluster_s.append(AgglomerativeClustering(n_clusters=clus[i], 
                                  affinity='precomputed',
                                  linkage='single'))  
    
        cluster_s[i].fit_predict(pmt[i])
        cluster_s[i].labels_=cluster_s[i].labels_+1
        unsupS.append(sklearn.metrics.adjusted_rand_score(pred_actual[i], cluster_s[i].labels_))


        cluster_spec.append(SpectralClustering(n_clusters=clus[i],assign_labels="discretize",random_state=0))
        cluster_spec[i].fit_predict(pmt[i])
        cluster_spec[i].labels_=cluster_spec[i].labels_+1
        unsupSpec.append(sklearn.metrics.adjusted_rand_score(pred_actual[i], cluster_spec[i].labels_))


        #cluster_ap.append(AffinityPropagation(n_clusters=clus[i]))
        #cluster_ap[i].fit_predict(pmt[i])
        #cluster_ap[i].labels_=cluster_ap[i].labels_+1
        #unsupAp.append(sklearn.metrics.adjusted_rand_score(pred_actual[i], cluster_ap[i].labels_))

        cluster_h.append(AgglomerativeClustering(n_clusters=clus[i]))
        cluster_h[i].fit_predict(pmt[i])
        cluster_h[i].labels_=cluster_c[i].labels_+1
        unsupH.append(sklearn.metrics.adjusted_rand_score(pred_actual[i], cluster_h[i].labels_))
    
        
        


    return unsupK,unsupC,unsupS,unsupSpec,unsupH#,unsupAp
```

```python
def data_prep_with_ratios(dr,arr):
    DATA = []
    LBL = []
    for dsm in dr:
        SmDt,lbl= dataPrep(dsm,arr)
        DATA.append(SmDt)
        LBL.append(lbl)
    return DATA,LBL
```

```python

def LeafWeightage_with_ratios(RF,dr,DATA,len_of_c_df):
    TL = []
    for i in range(0,len(dr)):
        tl=LeafWeightage(RF,DATA[i],len_of_c_df)
        TL.append(tl)
    return TL
```

```python

def getProx_with_ratios(LBL,TL,dr,len_ofc_df):
    PMT = []
    for i in range(0,len(dr)):
        pmt=getProx(LBL[i],TL[i],len_ofc_df)
        PMT.append(pmt)
    return PMT
```

```python

def getClus_with_ratios(PMT,dr,len_of_c_df):
    unsupKdr = []
    unsupCdr = []
    unsupSdr = []
    unsupSpecdr = []
    unsupW = []
    
    for i in range(0,len(dr)):
        a,b,c,d,e= get_clus(PMT[i],len_of_c_df)
        unsupKdr.append(a)
        unsupCdr.append(b)
        unsupSdr.append(c) 
        unsupSpecdr.append(d)
        unsupW.append(e)
    return unsupKdr,unsupCdr ,unsupSdr ,unsupSpecdr, unsupW 
    
```

```python

from sklearn.cluster import AffinityPropagation
from sklearn import metrics
from sklearn.datasets import make_blobs

# #############################################################################
# Generate sample data
centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(n_samples=300, centers=centers, cluster_std=0.5,
                            random_state=0)

# #############################################################################
# Compute Affinity Propagation
af = AffinityPropagation(preference=-50).fit(X)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_

n_clusters_ = len(cluster_centers_indices)

print('Estimated number of clusters: %d' % n_clusters_)
print("Homogeneity: %0.3f" % metrics.homogeneity_score(labels_true, labels))
print("Completeness: %0.3f" % metrics.completeness_score(labels_true, labels))
print("V-measure: %0.3f" % metrics.v_measure_score(labels_true, labels))
print("Adjusted Rand Index: %0.3f"
      % metrics.adjusted_rand_score(labels_true, labels))
print("Adjusted Mutual Information: %0.3f"
      % metrics.adjusted_mutual_info_score(labels_true, labels))
print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(X, labels, metric='sqeuclidean'))
```
